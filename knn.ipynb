```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python

X = np.array([[1, 2], [2, 3], [3, 1], [6, 5], [7, 7], [8, 6]])
Y = np.array([0, 0, 0, 1, 1, 1])
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='coolwarm')
plt.title('Data points and their classes')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()

```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, Y)
```

```python
new_X = np.array([[5, 5]])
predicted_class = knn.predict(new_X)
print("Predicted Class:", predicted_class[0])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='coolwarm', label="Existing points")

plt.scatter(new_X[:, 0], new_X[:, 1], color='green', marker='*', s=200, label="New point")

plt.title('KNN Classification')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
X[0] = [8, 8]
X[1] = [1, 1]
knn.fit(X, Y)

```

```python
Y[0] = 1
Y[1] = 0
knn.fit(X, Y)
predicted_class = knn.predict(new_X)
print("Predicted class after modification:", predicted_class[0])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='coolwarm', label="Existing points")
plt.scatter(new_X[:, 0], new_X[:, 1], color='green', marker='*', s=200, label="New point")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, Y)
predicted_class = knn.predict(new_X)

print("Predicted class with K=5:", predicted_class[0])
```

```python
new_X = np.array([[2, 2]])
predicted_class = knn.predict(new_X)

print("Predicted class for new point (2,2):", predicted_class[0])
```

```python
X = np.vstack([X, [6, 6]]) 
Y = np.append(Y, 2)

knn.fit(X, Y)

predicted_class = knn.predict(new_X)

print("Predicted class after adding new class:", predicted_class[0])
```
