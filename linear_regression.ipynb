```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
X = np.array([[15], [18], [21], [24], [27], [30], [33], [36], [39], [42]])
Y = np.array([2.5, 3.0, 3.5, 4.0, 4.6, 5.0, 5.5, 6.0, 6.5, 7.0])
```

```python
plt.scatter(X, Y, color='blue', label="Input Data")
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Ice cream price ($)")
plt.title("Dependence of ice cream price on temperature")
plt.legend()
plt.show()
```

```python
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, color='blue', label="Input Data")   
plt.plot(X, Y_pred, color='red', linewidth=2, label="The prediction line")   
plt.xlabel("temperature (Â°C)")
plt.ylabel("price of ice cream ($)")
plt.title("Linear regression: temperature vs price of ice cream")
plt.legend()
plt.show()
```

```python
new_temperature = np.array([[28]])
predicted_price = model.predict(new_temperature)
print("At 28Â°C, the predicted price of ice cream is: $", round(predicted_price[0], 2))
```
